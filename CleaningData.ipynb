```python
import pandas as pd
```

```python
# Load the data into a DataFrame 
df = pd.read_csv('mock_fashion_data_uk_us.csv')
df
```

```python
# Define the columns to drop
columns_to_drop = ['Social Media Comments', 'Fashion Influencers', 'Fashion Magazines', 'feedback', 'Product Name']

# Drop the specified columns
df.drop(columns=columns_to_drop, inplace=True)

df
```

```python
# Drop all NaN values if any
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
# Check if any NaN values are still present
df.isnull().sum()
```

```python
# Ensure price is of a numeric type 
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
# Normalizing the rating column
df['Rating'] = (df['Rating'] - df['Rating'].min()) / (df['Rating'].max() - df['Rating'].min()) * 5
```

```python
# One-Hot encoding for categorial variables 
columns_to_encode = ['Total Sizes', 'Available Sizes','Brand', 'Category', 'Style Attributes', 'Color', 'Purchase History', 'Season', 'Time Period Highest Purchase', 'Customer Reviews']

# Perform one-hot encoding using pd.get_dummies
df= pd.get_dummies(df, columns=columns_to_encode)
(df.head())
```

```python
# Ensuring text is properly readable for ML Model
df['Description'] = df['Description'].str.lower().str.replace('[^\w\s]', '', regex=True)
```

```python
# Checking dataframe
df.head()
```

```python
# Saving cleaned csv
cleaned_csv_path = 'cleaned_data.csv'  
df.to_csv(cleaned_csv_path, index=False)
```
